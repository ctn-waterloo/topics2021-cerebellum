```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
T = 0.5
dt = 1e-4
ts = np.arange(0, T, dt)
```

```python
tau1 = 33e-3
tau2 = 169e-3
flt1 = np.exp(-ts / tau1)
flt2 = np.exp(-ts / tau2)
ratio = 0.5

tau3 = 60e-3
flt3 = np.exp(-ts / tau3)
```

```python
fig, ax = plt.subplots()
im = plt.imread("dieudonne_nmda.png")
ax.imshow(im, extent=(-70e-3, 526e-3, -0.15, 1.5))
ax.set_aspect(0.3)
ax.plot(ts, ratio * flt1 + (1.0 - ratio) * flt2, linewidth=3)
ax.plot(ts, flt3, linewidth=3)
```
